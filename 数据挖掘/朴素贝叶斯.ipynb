# 计算概率

## 计算先验概率

```python
priori = {}
for features_name in features_list:
    #判断特征名是否在先验字典中
    if features_name not in priori.keys():
        priori[features_name]={}
    classCount = 0
    #得到个数
    for feature in data[features_name]:
        if feature not in priori[features_name].keys():
            priori[features_name][feature]=0
            classCount += 1
        priori[features_name][feature] += 1
    #拉普拉斯平滑
    for feature in priori[features_name].keys():
        priori[features_name][feature] = (priori[features_name][feature] + 1 ) /( 17 + classCount )
```

```python
#输出各个特征的先验概率
for features_name in features_list:
    print(features_name , priori[features_name])
```

## 计算后验概率

```python
#计算后验概率
features_list = np.array( data.columns)[0:6]
```

```python
good_poster = {}
for features_name in features_list:
    #判断特征名是否在先验字典中
    if features_name not in good_poster.keys():
        good_poster[features_name]={}
    classCount=0
    for feature in good[features_name]:
        if feature not in good_poster[features_name].keys():
            good_poster[features_name][feature]=0
            classCount +=1
        good_poster[features_name][feature] += 1 
    #检测是否有0值特征
    for feature in priori[features_name]:
        if feature not in good_poster[features_name].keys():
            classCount+=1
            good_poster[features_name][feature]=0
    #拉普拉斯平滑
    for feature in good_poster[features_name].keys():
        good_poster[features_name][feature] = (good_poster[features_name][feature] + 1 ) /( 8 + classCount )

for features_name in features_list:
    print(features_name , good_poster[features_name])
```

```python
bad_poster = {}
for features_name in features_list:
    #判断特征名是否在先验字典中
    if features_name not in bad_poster.keys():
        bad_poster[features_name]={}
    classCount =0
    for feature in bad[features_name]:
        if feature not in bad_poster[features_name].keys():
            bad_poster[features_name][feature]=0
            classCount+=1
        bad_poster[features_name][feature] += 1
    #检测是否有0值特征
    for feature in priori[features_name]:
        if feature not in bad_poster[features_name].keys():
            classCount+=1
            bad_poster[features_name][feature]=0
    #  拉普拉斯平滑
    for feature in bad_poster[features_name].keys():
        bad_poster[features_name][feature] = (bad_poster[features_name][feature] + 1 ) /( 9 + classCount )

for features_name in features_list:
    print(features_name , bad_poster[features_name])
```

```python
posterior = {'是':good_poster ,'否':bad_poster}
```

# 预测

P(Lk | Ak Bk Ck Dk Ek Fk) = P(Ak|Lk)P(Bk|Lk)P(Ck|Lk)P(Dk|Lk)P(Ek|Lk)P(Fk|Lk)*P(Lk)/(P(Ak)P(Bk)P(Ck)P(Dk)P(Ek)P(Fk))

```python
label_list = ['是','否']
```

```python
#根据上面的公式利用先验概率与后验概率计算条件概率
def contingent_probability( lable_id,color, gendi, sound, texture, qibu ,touch):
    p0 = posterior[label_list[lable_id]][features_list[0]][color] 
    p1 = posterior[label_list[lable_id]][features_list[1]][gendi] 
    p2 = posterior[label_list[lable_id]][features_list[2]][sound] 
    p3 = posterior[label_list[lable_id]][features_list[3]][texture] 
    p4 = posterior[label_list[lable_id]][features_list[4]][qibu]
    p5 = posterior[label_list[lable_id]][features_list[5]][touch]
    P = priori['好瓜'][label_list[0]] 
    P0 = priori[features_list[0]][color]
    P1 = priori[features_list[1]][gendi]
    P2 = priori[features_list[2]][sound]
    P3 = priori[features_list[3]][texture]
    P4 = priori[features_list[4]][qibu]
    P5 = priori[features_list[5]][touch] 
    
    return  p0*p1*p2*p3*p4*p5*P/(P0*P1*P2*P3*P4*P5)
```

```python
#朴素贝叶斯分类器
def NB_Classifier(color, gendi, sound , texture , qibu,touch):
    good_prob = contingent_probability(0,color, gendi, sound, texture, qibu ,touch)
    bad_prob = contingent_probability(1,color, gendi, sound, texture, qibu ,touch)
    if(good_prob > bad_prob):
        out = '是'
    else:
        out = '否'
    return out
```

# 在验证集上测试

```python
valData = pandas.read_csv('西瓜的验证集.csv')
dataV  = valData.values
```

```python
acc=[]
print('测试结果\t真实结果')
for feature in dataV:
    color = feature[0]
    gendi = feature[1]
    sound = feature[2]
    texture = feature[3]
    qibu = feature[4]
    touch = feature[5]
    label = feature[6]
    out = NB_Classifier(color,gendi,sound,texture,qibu,touch)
    print(out ,'\t', label)
    acc.append(out==label)
print("正确率为:" , sum(acc)/len(acc))
```
