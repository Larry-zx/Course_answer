## 导入相关库

```python
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd 
```

## 分析数据格式

```python
# 准备数据集
iris=load_iris()
```

```python
#了解数据集
print('字典包含以下关键字:')
for key in iris:
    print(key)
```

```python
#label
print(iris['target_names'])
#feature
print(iris['feature_names'])
#data
print(iris['data'])
```

```python
#获取四个特征向量数据
sepal_length = iris['data'][:,0]
sepal_width = iris['data'][:,1]
petal_length = iris['data'][:,2]
petal_width = iris['data'][:,3]
```

## 直方图

```python
plt.subplot(2,2,1)
plt.hist(sepal_length)
plt.title("sepal_length")
plt.subplot(2,2,2)
plt.hist(sepal_width)
plt.title("sepal_width")
plt.subplot(2,2,3)
plt.hist(petal_length)
plt.title("petal_length")
plt.subplot(2,2,4)
plt.hist(petal_width)
plt.title("petal_width")
plt.show()
```

## 散点图
> 散点图的成对关系有12种 这里选了两种进行展示

```python
#sepal_length与sepal_width之间的散点图
plt.scatter(sepal_length, sepal_width,marker='o')
plt.show()
```

```python
#petal_length与petal_width之间的散点图
plt.scatter(petal_length, petal_width,marker='x')
plt.show()
```

## 箱线图

```python
#matplotlib
plt.boxplot(iris['data'],labels=iris['feature_names'])
plt.show()
```

```python
#sns
import seaborn as sns
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
sns.boxplot(data=df)
plt.show()
```

## 平行坐标

```python
import seaborn as sns
from pandas.plotting import parallel_coordinates #导入平行坐标的包
iris_sns = sns.load_dataset('iris')
fig,axes = plt.subplots()
parallel_coordinates(iris_sns,'species',ax=axes)
```
